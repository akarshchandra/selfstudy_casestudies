# OBJECTIVE

##### Classifing wether wether the patient survives 5 years or more  based on age,year,auxillary nodes 

### Univariate Analysis:-
####  Histogram of the independent features :-

```python
Hdata_1=Hdata[Hdata["survival"]==1]
Hdata_2=Hdata[Hdata["survival"]==2]
plt.close()
sns.set_style("whitegrid");
sns.FacetGrid(Hdata,hue="survival",size=4).map(sns.distplot,'Age').add_legend();
plt.show()
```

```python
plt.close()
sns.set_style("whitegrid");
sns.FacetGrid(Hdata,hue="survival",size=4).map(sns.distplot,"Year").add_legend()
plt.show()
```

```python
plt.close()
sns.set_style("whitegrid");
sns.FacetGrid(Hdata,hue="survival",size=9).map(sns.distplot,"axilnodes").add_legend()
plt.show()
```

## Observation :-

###### We can see that it is not possible to clasify wether the patient survives more than 5 years or less than five yers by looking into each independent variables as Histogram of each feature there is an overlap of the survival status 1 & 2  and hence Single feature alone does not seem useful for clasification 



## PDF  and CDF 

```python
count,binedges=np.histogram(Hdata_1["Age"],bins=10,density=True)
pdf=count/(sum(count))
cdf=np.cumsum(pdf)
plt.plot(binedges[1:],pdf)
plt.plot(binedges[1:],cdf)

count,binedges=np.histogram(Hdata_2["Age"],bins=10,density=True)
pdf=count/(sum(count))
cdf=np.cumsum(pdf)
plt.plot(binedges[1:],pdf)
plt.plot(binedges[1:],cdf)
```

###### Observation:-

1.CDF of both survival>=5 (i.e survial=1) and survival<5 i.e (survival =2) almost ovrerlap 
2.20% of patients  who survived 5 years or more were operated at or less than 41 years of age
  20% of the patients who didnt survive for 5 years got operated at or less than 45 years of age 


```python
plt.close()
cnt,bins=np.histogram(Hdata_1["Year"],density=True)
pdf=cnt/(sum(cnt))
cdf=np.cumsum(pdf)
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)


cnt,bins=np.histogram(Hdata_2["Year"],density=True)
pdf=cnt/(sum(cnt))
cdf=np.cumsum(pdf)
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)
plt.show()
```

```python
plt.close()
cnt,bins=np.histogram(Hdata_1["axilnodes"],density=True)
pdf=cnt/(sum(cnt))
cdf=np.cumsum(pdf)
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)


cnt,bins=np.histogram(Hdata_2["axilnodes"],density=True)
pdf=cnt/(sum(cnt))
cdf=np.cumsum(pdf)
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)
plt.show()
```

```python
plt.close()
sns.boxplot(x="survival",y="Age",data=Hdata)
plt.show()
plt.close()
sns.boxplot(x="survival",y="Year",data=Hdata)
plt.show()
plt.close()
sns.boxplot(x="survival",y="axilnodes",data=Hdata)
plt.show()
```

#### Observations :-

1.Box plots for Age fig(1) are almost similar in terms of ages of 25,50,75 percentile

2.Box plots for Year fig(2) are almost similar except 25 % of operations of patients who survived more than 5 years happened in 1960 or later and 75 % occoured before 1966 wheras 25 % of patients who did not survive 5 years were operated before 1959 and 75% before 1965 

3.Box plots for auxillary nodes fig(3) 75% of patients who survived for greater than or equal to 5 years had less than 2 positive auxillary nodes wheras 75%  of patients who did not survive for 5 years had less than 11 positive auxillary nodes
